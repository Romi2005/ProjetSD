## <span style="color:blue">Fonction pour lire un fichier csv</span>

```python
def extraction_csv(repertoire_vise,fichier_source,separateur,decimale="."):
    # utiliser la fonctionread_csv de panda
    trame = pd.read_csv(os.path.join(repertoire_vise,fichier_source), sep=separateur, decimal=decimale)
    return trame
```

## <span style="color:blue">Préparation des données en vue de l'insertion dans la base de données primaire</span>

### Préparation des données de pollution

```python
# Dans un premier tant récupérons la colonne d'intérêt

trame_pollution_final = pd.DataFrame()

for i in tqdm(range(len(trame_pollution_list))) :
    trame = []
    
    colonne = trame_pollution_list[i].columns[1]
    # Récupérons les données d'intérêt de la colonne 
    nomVar = colonne.split(",")
     #print("Fichier",i+1,":",nomVar)
     #print("Nombre de données en entré", len(trame_pollution_list[i]))
    
    for j in range(len(trame_pollution_list[i])) :
    
        if len(nomVar) > 1 :
            trame += [[trame_pollution_list[i]["DateTime"][j], nomVar[0], nomVar[1].replace(" ",""),  trame_pollution_list[i][colonne][j]]]
        
        else : # La structure de la dernière colonne est différente 16, 25 et 30 
            nomVar = nomVar[0].split(" ")
            trame += [[trame_pollution_list[i]["DateTime"][j], nomVar[0], nomVar[1].replace(" ",""),  trame_pollution_list[i][colonne][j]]]
            
    # Convertisons la liste en un objet pandas 
    dataframe = pd.DataFrame(trame, columns = ["Date", "station", "typeMesure", "valeurMesure"])
        
        
     #print("Nombre de données en sortie", len(dataframe))
     #print("Une description de notre dataframe des données : \n",dataframe.describe, "\n")

    trame_pollution_final = pd.concat([trame_pollution_final,dataframe], ignore_index = True)
    
trame_pollution_final.describe
        
    
```

### Préparation des données de météo

```python
trame_meteo_final = pd.DataFrame()

for i in range(len(trame_meteo_list)) :
    new_trame_meteo = []
    
    # Renommer les colonnes
    
    # Trame à trois colonnes
    if i == 0:
        trame_meteo_list[i] = trame_meteo_list[i].rename(columns = {'DateTime' : 'Date',
                                                                    "Max Wind Speed 10 m (m/s)" : "MaxWindSpeed10",
                                                                    "Max Wind Speed 24 m (m/s)" : "MaxWindSpeed24"
                                                                   })

    elif i == 1:
        trame_meteo_list[i] = trame_meteo_list[i].rename(columns = {'DateTime' : 'Date',
                                                                    "DiffTemp 2-8 m (deg.C)" : "DiffTemp2_8",
                                                                    "DiffTemp 8-22 m (deg.C)" : "DiffTemp8_22"})
    elif i == 4:
        trame_meteo_list[i] = trame_meteo_list[i].rename(columns = {'DateTime' : 'Date',
                                                                    "Wind Direction 10 m (deg.M)" : "WindDirection10",
                                                                    "Wind Direction 24 m (deg.M)"  :  "WindDirection24"})
    elif i == 6:
        trame_meteo_list[i] = trame_meteo_list[i].rename(columns = {'DateTime' : 'Date',
                                                                    "Wind Speed 10 m (m/s)" : "WindSpeed10",
                                                                    "Wind Speed 24 m (m/s)" : "WindSpeed24"
                                                                   })
    # Trame à deux colonnes
    elif i == 2:
        trame_meteo_list[i] = trame_meteo_list[i].rename(columns = {'DateTime' : 'Date',
                                                                    "Athletic Stadium, Pressure, 000[M], Value (hPa)" : "Pressure"})
    elif i == 3:
        trame_meteo_list[i] = trame_meteo_list[i].rename(columns = {'DateTime' : 'Date',
                                                                    "Athletic Stadium, Rain, 000[M], Value (mm)" : "Rain"})
    elif i == 5 :
        trame_meteo_list[i] = trame_meteo_list[i].rename(columns = {'DateTime' : 'Date',
                                                                    "Athletic Stadium, Temp, 002[M], Value (deg.C)" : "Temp"})
    else:
        trame_meteo_list[i] = trame_meteo_list[i].rename(columns = {'DateTime' : 'Date',
                                                                    "Athletic Stadium, Global Radiation, 000[M], Value (W/m2)" : "Radiation"})
    
    # Vérification colonnes renommées
    print(trame_meteo_list[i].columns)

    new_trame_meteo = pd.melt(trame_meteo_list[i],
                                 id_vars = "Date",
                                 var_name = "typeMesure",
                                 value_name = "valeurMesure",
                                 ignore_index = True)
        
        
    trame_meteo_final = pd.concat([trame_meteo_final,new_trame_meteo], ignore_index = True)

trame_meteo_final["station"] = "Sheffield"

# Ordonnons les champs dans le même ordre que celui des données de pollution
trame_meteo_final = trame_meteo_final[["Date", "station", "typeMesure", "valeurMesure"]]

trame_meteo_final.describe
```
